# merge

두 개의 DataFrame을 공통된 컬럼(들)을 기준으로 합치는 것. SQL의 join과 같은 기능.

```python
emp = pd.DataFrame(data={
    'empno': [100, 101, 200, 201],
    'ename': ['Scott', 'King', 'Allen', 'Tiger'],
    'deptno': [10, 20, 10, 50]
})
```

```python
emp
```

```python
dept = pd.DataFrame(data={
    'deptno': [10, 20, 30],
    'dname': ['HR', 'IT', 'Sales']
})
```

```python
dept
```

merge(join)

*  `pd.merge(df1, df2, how, on, ...)` 함수
*  `pd.DataFrame.merge(df, how, on, ...)` 메서드
*  `how` 파라미터: merge(join) 방식. inner, left, right, outer
    *  merge 방식의 기본값은 inner. `how=inner`는 생략 가능
*  `on` 파라미터: merge(join)의 기준이 되는 컬럼(컬럼들의 리스트).
    *  merge하려는 2개의 데이터 프레임에서 컬럼 이름이 같은 경우에는 on 파라미터를 생략할 수 있음.

```python
pd.merge(emp, dept, how='inner', on='deptno')
# how='inner' 생략 가능
# emp와 dept에서 deptno 컬럼 이름이 동일. -> on 생략 가능
```

```python
emp.merge(dept)
```

## left (outer) join

```python
pd.merge(emp, dept, how='left')
```

## right (outer) join

```python
pd.merge(emp, dept, how='right')
```

## (full) outer join

```python
pd.merge(emp, dept, how='outer')
```

## join의 기준이 되는 컬럼 이름이 다른 경우

```python
employees = pd.DataFrame(data={
    'empno': [100, 101, 200, 201],
    'ename': ['Scott', 'King', 'Allen', 'Tiger'],
    'deptno': [10, 20, 10, 50]
})
employees
```

```python
departments = pd.DataFrame(data={
    'dno': [10, 20, 30],
    'dname': ['HR', 'IT', 'Sales']
})
departments
```

```python
pd.merge(employees, departments, left_on='deptno', right_on='dno')
```

## 인덱스를 사용한 merge

```python
np.random.seed(1)

df1 = pd.DataFrame(data={'val_1': np.random.randint(10, size=4),
                         'val_2': np.random.randint(100, size=4)},
                   index=['a', 'b', 'c', 'd'])
df1
```

```python
df2 = pd.DataFrame(data={'data_1': np.random.randn(4),
                         'data_2': np.random.randn(4)},
                   index=['d', 'b', 'c', 'a'])
df2
```

```python
pd.merge(df1, df2, left_index=True, right_index=True)
```

## 인덱스와 컬럼을 사용한 merge

```python
df3 = pd.DataFrame(data={'name': ['d', 'b', 'c', 'a'],
                         'data': np.random.randint(100, size=4)})
df3
```

```python
pd.merge(df1, df3, left_index=True, right_on='name').reset_index(drop=True)
# reset_index(): 인덱스를 데이터 프레임의 컬럼으로 만들고, 인덱스를 리셋.
# reset_index(drop=True): 기존 인덱스를 삭제하고 리셋.
```

```python
pd.merge(df1, df3, left_index=True, right_on='name').set_index('name')
# set_index(col_name): col_name을 인덱스로 설정.
```

# concat

`pd.Series` 또는 `pd.DataFrame`을 축(axis)를 따라서 합치는 방법.

## Series에서의 concat

```python
s1 = pd.Series(data=['a', 'b', 'c'])
s1
```

```python
s2 = pd.Series(data=['d', 'e'])
s2
```

```python
pd.concat([s1, s2], axis=0)
# axis=0은  생략 가능(default argument).
```

```python
pd.concat([s1, s2], ignore_index=True)
```

```python
pd.concat([s1, s2], axis=1)
#> Series 2개를 axis=1 방향으로 합치면 그 결과는 DataFrame.
```

## DataFrame에서의 concat

```python
np.random.seed(1)

df1 = pd.DataFrame(data={'id': np.arange(1, 6),
                         'kor': np.random.randint(100, size=5),
                         'eng': np.random.randint(100, size=5)})
df1
```

```python
df2 = pd.DataFrame(data={'id': np.arange(6, 11),
                         'kor': np.random.randint(100, size=5),
                         'eng': np.random.randint(100, size=5)})
df2
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)  # axis=0은 생략 가능.
```

```python
df3 = pd.DataFrame(data={'math': np.random.randint(100, size=5),
                         'sci': np.random.randint(100, size=5),
                         'hist': np.random.randint(100, size=5)})
df3
```

```python
pd.concat([df1, df3], axis=1)
```

# babynames 데이터 셋


*   https://github.com/wesm/pydata-book/tree/2nd-edition/datasets/babynames 데이터 파일들을 이용.
*   1880년부터 2010년까지 미국 신생아들의 이름,성별,출생자수가 정리된 파일들.
*   파일을 하나씩 읽어서 데이터 프레임을 생성.
*   각 데이터 프레임에 연도 컬럼을 추가.
*   생성된 131개의 데이터 프레임을 하나로 합침.
*   연도별 신생아 숫자를 찾고 시각화.
*   연도별 성별 신생아 숫자를 찾고 시각화.


```python
yobs = []  # 각 연도의 데이터 프레임들을 저장할 리스트(year-of-births)
col_names = ['name', 'gender', 'births']  # 데이터 프레임의 컬럼 이름
for year in range(1880, 2011):  # 1880 ~ 2010년까지 반복하면서
    # txt 파일이 저장된 경로
    file_path = f'https://github.com/wesm/pydata-book/raw/2nd-edition/datasets/babynames/yob{year}.txt'
    
    # 파일을 읽어서 DataFrame 생성
    df = pd.read_csv(file_path, names=col_names)
    
    # DataFrame 연도(year) 컬럼을 추가
    df['year'] = year

    # DataFrame을 리스트에 추가
    yobs.append(df)
```

```python
len(yobs)
```

```python
yobs[0]  #> 1880년 DataFrame
```

```python
yobs[-1]  #> 2010년의 DataFrame
```

```python
# 131개의 DataFrame을 이어붙이기(concat)
yob_df = pd.concat(yobs)
```

```python
yob_df.info()
```

```python
yob_df.head()
```

```python
yob_df.tail()
```

```python
# 연도별 신생아 숫자
births_by_year = yob_df.groupby('year')['births'].sum()
births_by_year  #> Series
```

```python
# groupby 결과 Series의 인덱스를 컬럼으로 reset
births_by_year = births_by_year.reset_index()
births_by_year  #> DataFrame
```

```python
sns.lineplot(x='year', y='births', data=births_by_year)
plt.show()
```

```python
# 연도별 성별 신생아 수
births_by_year_gender = yob_df.groupby(by=['year', 'gender'])['births'].sum()
births_by_year_gender  #> MultiIndex를 갖는 Series
```

```python
births_by_year_gender.reset_index()
# level 파라미터를 지정하지 않으면, MultiIndex 모두 컬럼으로 변환.
```

```python
births_by_year_gender.reset_index(level=1)
# level을 지정하면 지정한 인덱스만 컬럼으로 변환.
# level의 순서는 0, 1, 2, ...
```

```python
births_by_year_gender.reset_index(level='gender')
```

```python
births_by_year_gender = births_by_year_gender.reset_index()
births_by_year_gender
```

```python
sns.lineplot(x='year', y='births',hue='gender', data=births_by_year_gender)
plt.show()
```

```python
stations[stations.duplicated('location', keep='last')]
#> 중복된 값이 있을 때 처음에 나오는 값을 중복된 값으로 처리, 나중에 나오는 값은 유지
```

```python
stations[stations.duplicated('location', keep=False)]
#> 중복된 모든 값을 중복된 값으로 처리.
```
