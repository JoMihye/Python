```python
import pandas as pd
```

```python
# GitHub에 저장된 CSV 파일 경로
ex1_csv = 'https://github.com/wesm/pydata-book/raw/2nd-edition/examples/ex1.csv'
```

```python
ex1 = pd.read_csv(ex1_csv)
```

```python
ex1
```

```python
ex1.info()
```

`pd.read_csv()` 함수:

*  CSV 파일의 첫번째 줄은 데이터가 아니라 컬럼 이름으로 간주.
*  값과 값을 구분하는 구분자의 기본값은 쉼표(,).
*  변수 타입을 숫자로 변환할 수 있는 경우에는 자동으로 변환을 수행함.

# `pd.read_excel()`

```python
ex1_xlsx = 'https://github.com/wesm/pydata-book/raw/2nd-edition/examples/ex1.xlsx'
```

```python
ex1 = pd.read_excel(ex1_xlsx)
```

```python
ex1
```

```python
ex1.info()
```

`pd.read_excel()` 함수

*  `read_csv()`와 마찬가지로 첫번째 행은 컬럼 이름으로 간주.
*  각 컬럼의 데이터 타입도 자동으로 변환(int, float, object)

```python
# 엑셀 파일에서 index 컬럼(행 번호 또는 레이블들을 가지고 있는 컬럼)이 있는 경우,
# index 컬럼을 실제 데이터에서 제거하기 위해서
ex1 = pd.read_excel(ex1_xlsx, index_col=0)
ex1
```

# 컬럼 이름이 없는 csv 파일

```python
ex2_csv = 'https://github.com/wesm/pydata-book/raw/2nd-edition/examples/ex2.csv'
```

```python
ex2 = pd.read_csv(ex2_csv)
ex2
```

```python
ex2.info()
```

ex2.csv 파일에는 컬럼 이름들이 작성되어 있지 않고, 첫번째 줄부터 데이터가 시작.

```python
ex2 = pd.read_csv(ex2_csv, names=['x1', 'x2', 'x3', 'x4', 'msg'])
ex2
```

```python
ex2.info()
```

# 구분자(seperator, delimeter)

```python
ex3_csv = 'https://github.com/wesm/pydata-book/raw/2nd-edition/examples/ex3.csv'
```

```python
ex3 = pd.read_csv(ex3_csv, sep=r'\s+')
# sep=r'\s+': 구분자가 한 개 이상의 공백(space)임을 지정. 정규 표현식(regular expression)
ex3
```

```python
ex3.info()
```

`pd.read_csv()` 함수는 csv(txt, tsv) 파일에서 이름이 없는 컬럼은 row index(레이블)로 간주.

```python
ex3.loc['aaa':'bbb']
```

```python
ex3.iloc[0:2]
```

# 문서에 대한 설명(주석)을 가지고 있는 csv 파일

```python
ex4_csv = 'https://github.com/wesm/pydata-book/raw/2nd-edition/examples/ex4.csv'
```

```python
ex4 = pd.read_csv(ex4_csv, skiprows=[0, 2, 3])
# skiprows: 생략할 행 번호들의 리스트 또는 생략할 행의 개수. 행 번호는 0부터 시작.
ex4
```
